**Langkah 6: Regresi Linier**


```python
# Membuat variabel bebas X dan Y, contoh pengambilan dari analisis korelasi sebelumnya
X = data['Length of Membership']
y = data['Yearly Amount Spent']
```

*   Pisahkan variabel bebas (X) dan variabel target (y).

**Penjelasan:**

Kode tersebut membuat dua variabel, X dan y, yang umumnya digunakan dalam pemodelan statistik.
* **X = data['Length of Membership']:**
    - Membuat variabel X dan mengisinya dengan nilai dari kolom 'Length of Membership' dari DataFrame data.
    - Kegunaan: Dalam konteks machine learning, X biasanya merepresentasikan variabel independen atau fitur yang digunakan untuk memprediksi variabel dependen. 'Length of Membership' diasumsikan sebagai fitur yang mungkin berpengaruh terhadap jumlah uang yang dihabiskan per tahun.

* **y = data['Yearly Amount Spent']:**
    - Membuat variabel y dan mengisinya dengan nilai dari kolom 'Yearly Amount Spent' dari DataFrame data.
    - Kegunaan: Dalam konteks machine learning, y biasanya merepresentasikan variabel dependen atau target yang ingin diprediksi. 'Yearly Amount Spent' adalah variabel yang ingin diprediksi berdasarkan 'Length of Membership'.

```python
# Pembagian data latih dan data uji dengan proporsi 7:3
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

*   Bagi data menjadi data latih (70%) dan data uji (30%) menggunakan train_test_split.
*   Lakukan training model regresi linier menggunakan library StatsModels. Tambahkan konstanta (intercept) ke variabel bebas.

**Penjelasan:**

Kode tersebut membagi data menjadi data latih (training data) dan data uji (testing data) dengan proporsi 70% untuk data latih dan 30% untuk data uji.
* **from sklearn.model_selection import train_test_split:**
  - Mengimpor fungsi train_test_split dari modul sklearn.model_selection. Fungsi ini digunakan untuk membagi dataset menjadi subset data latih dan data uji.
* **X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100):**
  - Memanggil fungsi train_test_split untuk membagi data.
  - X: Variabel independen (fitur) yang telah didefinisikan sebelumnya.
  - y: Variabel dependen (target) yang telah didefinisikan sebelumnya.
  - train_size=0.7: Menentukan proporsi data latih sebesar 70%.
  - test_size=0.3: Menentukan proporsi data uji sebesar 30%.
  - random_state=100: Menentukan seed untuk generator angka acak. Ini memastikan bahwa pembagian data akan sama setiap kali kode dijalankan, sehingga hasilnya dapat direproduksi.
  - X_train, X_test, y_train, y_test: Variabel untuk menyimpan hasil pembagian data, yaitu fitur dan target untuk data latih dan data uji.

```python
# Training model
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm).fit()
```

*   Visualisasikan garis regresi pada data latih.

**Penjelasan:**

Kode tersebut digunakan untuk melatih model regresi linier menggunakan library statsmodels.
* **import statsmodels.api as sm:** Mengimpor library statsmodels dan memberinya alias sm untuk mempermudah penggunaan. statsmodels adalah library Python yang digunakan untuk eksplorasi data, estimasi model statistik, dan pengujian hipotesis.
* **X_train_sm = sm.add_constant(X_train):**
  - Menambahkan konstanta (intercept) ke variabel independen (X_train).
  - Kegunaan: Dalam regresi linier, konstanta merepresentasikan nilai y ketika semua variabel independen bernilai nol. Menambahkan konstanta memungkinkan model untuk menyesuaikan intercept dari garis regresi. statsmodels membutuhkan konstanta untuk didefinisikan secara eksplisit.
* **lr = sm.OLS(y_train, X_train_sm).fit():**
  - Membuat dan melatih model regresi linier (Ordinary Least Squares - OLS).
  - **sm.OLS(y_train, X_train_sm):** Membuat objek model OLS dengan y_train sebagai variabel dependen dan X_train_sm sebagai variabel independen (termasuk konstanta).
  - **.fit():** Melatih model pada data latih. Metode ini mengestimasi parameter model (koefisien regresi) yang paling sesuai dengan data.
  - **lr:** Menyimpan model yang telah dilatih ke dalam variabel lr.

Jadi, setelah kode ini dijalankan, variabel lr akan berisi model regresi linier yang telah dilatih dan dapat digunakan untuk membuat prediksi pada data baru.

```python
# Visualisasi garis regresi
plt.scatter(X_train, y_train)
plt.plot(X_train, 265.2483 + 66.3015*X_train, 'r')
plt.show()
```

**Penjelasan:**

Kode tersebut memvisualisasikan garis regresi yang diperoleh dari model yang telah dilatih sebelumnya, diplot bersama dengan scatter plot data latih.
* **plt.scatter(X_train, y_train):** Membuat scatter plot dari data latih.
  - **X_train:** Variabel independen (fitur) dari data latih.
  - **y_train:** Variabel dependen (target) dari data latih.
  - **Kegunaan:** Scatter plot menampilkan titik-titik data, memungkinkan kita untuk melihat hubungan antara variabel independen dan dependen secara visual.

* **plt.plot(X_train, 265.2483 + 66.3015*X_train, 'r'):** Memplot garis regresi.
  - **X_train:** Variabel independen digunakan untuk menghasilkan nilai prediksi pada sumbu y.
  - **265.2483 + 66.3015*X_train:** Persamaan garis regresi. Nilai 265.2483 adalah intercept dan 66.3015 adalah koefisien regresi (slope). Nilai-nilai ini mungkin berbeda tergantung pada hasil training model Anda.
  - **'r':** Menentukan warna garis regresi sebagai merah.
  - **Kegunaan:** Garis regresi menunjukkan hubungan linier yang dimodelkan antara variabel independen dan dependen.
plt.show(): Menampilkan plot yang telah dibuat.

Dengan memvisualisasikan garis regresi bersama dengan scatter plot data latih, kita dapat melihat seberapa baik model menyesuaikan data dan memahami tren yang ditangkap oleh model.

**Langkah 7: Analisis Residual**

```python
# Prediksi nilai y_value dari data x yang telah dilatih
y_train_pred = lr.predict(X_train_sm)

# Menghitung residual
res = (y_train - y_train_pred)
```

*   Lakukan prediksi nilai y dari data latih dan hitung residual (selisih antara nilai sebenarnya dan nilai prediksi).

**Penjelasan:**

Kode tersebut memprediksi nilai variabel dependen (y) menggunakan model yang telah dilatih dan menghitung residual (selisih antara nilai aktual dan nilai prediksi).
* **y_train_pred = lr.predict(X_train_sm):**
  - Memprediksi nilai y (dalam hal ini, 'Yearly Amount Spent') untuk data latih menggunakan model regresi linier (lr) yang telah dilatih.
  - **lr.predict(X_train_sm):** Memanggil metode predict() pada objek model lr dengan X_train_sm (variabel independen data latih yang sudah ditambahkan konstanta) sebagai input.
  - **y_train_pred:** Menyimpan hasil prediksi ke dalam variabel y_train_pred.

* **res = (y_train - y_train_pred):**
  - Menghitung residual, yaitu selisih antara nilai aktual y_train dan nilai prediksi y_train_pred.
  - **res:** Menyimpan nilai residual ke dalam variabel res.

```python
# Histogram residual
fig = plt.figure()
sns.distplot(res, bins=15)
plt.title('Error Terms', fontsize=15)
plt.xlabel('y_train - y_train_pred', fontsize=15)
plt.show()

# Scatter plot residual
plt.scatter(X_train, res)
plt.show()
```

*   Visualisasikan residual dalam bentuk histogram dan scatter plot untuk mengevaluasi distribusi dan pola error

**Penjelasan:**

Kode tersebut membuat dua visualisasi untuk menganalisis residual dari model regresi linier: histogram dan scatter plot.

**1. Histogram Residual**
  * Membuat histogram dari residual (res).
  * **Kegunaan:** Histogram residual digunakan untuk memeriksa apakah residual berdistribusi normal. Distribusi normal residual merupakan salah satu asumsi penting dalam regresi linier.
  - **fig = plt.figure():** Membuat objek figure baru.
  - **sns.distplot(res, bins=15):** Membuat histogram menggunakan fungsi distplot dari Seaborn. bins=15 menentukan jumlah bin (interval) dalam histogram.
  - **plt.title('Error Terms', fontsize=15):** Menambahkan judul plot.
  - **plt.xlabel('y_train - y_train_pred', fontsize=15):** Menambahkan label pada sumbu x.

**2. Scatter Plot Residual**
  * Membuat scatter plot residual terhadap variabel independen (X_train).
  * **Kegunaan:** Scatter plot residual digunakan untuk:
      - Memeriksa apakah ada pola non-linier dalam residual. Jika ada pola non-linier, mungkin model regresi linier tidak tepat untuk data.
      - Mendeteksi heteroskedastisitas, yaitu varians residual yang tidak konstan. Heteroskedastisitas melanggar asumsi regresi linier.
      - Mengidentifikasi outlier, yaitu titik data dengan residual yang sangat besar.

**Langkah 8: Prediksi pada Data Uji dan Evaluasi Model**

```python
# Prediksi pada data uji
X_test_sm = sm.add_constant(X_test)
y_test_pred = lr.predict(X_test_sm)
```

*   Lakukan prediksi pada data uji.

**Penjelasan:**

Kode tersebut digunakan untuk melakukan prediksi pada data uji menggunakan model regresi linier yang telah dilatih sebelumnya.

* **X_test_sm = sm.add_constant(X_test):**
  * Menambahkan konstanta (intercept) ke variabel independen pada data uji (X_test).
  * **Kegunaan:** Sama seperti pada data latih, penambahan konstanta diperlukan agar model dapat memprediksi nilai y dengan benar.

* **y_test_pred = lr.predict(X_test_sm):**
  * Memprediksi nilai variabel dependen (y) pada data uji menggunakan model regresi linier (lr) yang telah dilatih sebelumnya.
  * **lr.predict(X_test_sm):** Memanggil metode predict() pada objek model lr dengan X_test_sm (data uji dengan konstanta) sebagai input.
  * **y_test_pred:** Menyimpan hasil prediksi pada data uji ke dalam variabel y_test_pred.

```python
# Hitung nilai R-squared
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, y_test_pred)
```

*   Hitung nilai R-squared untuk mengukur kinerja model pada data uji.

**Penjelasan:**

Kode tersebut menghitung nilai R-squared untuk mengevaluasi kinerja model regresi linier pada data uji.

  * Mengimpor fungsi r2_score dari modul sklearn.metrics. Fungsi ini digunakan untuk menghitung nilai R-squared.

* **r_squared = r2_score(y_test, y_test_pred):**
  * Memanggil fungsi r2_score dengan argumen:
    - **y_test:** Nilai aktual dari variabel dependen pada data uji.
    - **y_test_pred:** Nilai prediksi dari variabel dependen pada data uji.
    - **r_squared:** Menyimpan nilai R-squared yang dihitung ke dalam variabel r_squared.

  * Membaca data dari file CSV 'Posisi_gaji.csv' yang berada dalam folder 'dataset' dan menyimpannya ke dalam variabel dataset sebagai DataFrame Pandas.
  * **pd.read_csv():** Fungsi dari Pandas untuk membaca file CSV.

* **X = dataset.iloc[:, 1:2].values:**
  * Membuat variabel X yang berisi nilai-nilai dari kolom kedua (indeks 1) dataset.
  * **dataset.iloc[:, 1:2]:** Menggunakan iloc untuk melakukan slicing pada DataFrame.
    - : sebelum koma berarti memilih semua baris.
    - 1:2 setelah koma berarti memilih kolom dengan indeks 1 (kolom kedua). Penulisan 1:2 digunakan untuk menghasilkan array 2 dimensi, yang dibutuhkan oleh beberapa model machine learning.
  * **.values:** Mengambil nilai dari DataFrame dan mengubahnya menjadi array NumPy.

* **y = dataset.iloc[:, 2].values:**
  * Membuat variabel y yang berisi nilai-nilai dari kolom ketiga (indeks 2) dataset.
  * **dataset.iloc[:, 2]:** Memilih semua baris dan kolom dengan indeks 2.
  * **.values:** Mengubah nilai dari DataFrame menjadi array NumPy.

**3. Feature Scaling:**

  Menggunakan StandardScaler untuk melakukan penskalaan fitur X dan target y. Ini diperlukan karena SVM sangat sensitif terhadap skala data.

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X.reshape(-1, 1))
y = sc_y.fit_transform(y.reshape(-1, 1))
```

**Penjelasan:**

Kode tersebut melakukan feature scaling pada variabel X dan y menggunakan StandardScaler dari library Scikit-learn.

* **from sklearn.preprocessing import StandardScaler:** Mengimpor kelas StandardScaler dari modul sklearn.preprocessing.
* **sc_X = StandardScaler():** Membuat objek StandardScaler untuk variabel X.
* **sc_y = StandardScaler():** Membuat objek StandardScaler untuk variabel y.
* **X = sc_X.fit_transform(X.reshape(-1, 1)):**
    * **X.reshape(-1, 1):** Mengubah bentuk array X menjadi array 2 dimensi dengan satu kolom. Ini diperlukan karena StandardScaler mengharapkan input 2 dimensi. -1 berarti jumlah baris akan dihitung secara otomatis.
    * **sc_X.fit_transform(...):** Melakukan fitting dan transformasi pada X.
      - **fit():** Menghitung mean dan standar deviasi dari data.
      - **transform():** Mentransformasi data menggunakan mean dan standar deviasi yang telah dihitung.
* **y = sc_y.fit_transform(y.reshape(-1, 1)):** Melakukan hal yang sama seperti langkah 4, tetapi untuk variabel y.

**4. Fitting SVR ke Dataset:**

  Lakukan pembuatan model SVR dengan kernel RBF (Radial Basis Function) dan melatihnya dengan data yang telah di-scaled.

```python
# Fitting SVR ke dataset
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(X, y)
```

**Penjelasan:**

Kode tersebut membuat dan melatih model Support Vector Regression (SVR) dengan kernel RBF (Radial Basis Function) untuk dataset yang telah disiapkan.

* **from sklearn.svm import SVR:** Mengimpor kelas SVR dari modul sklearn.svm. SVR adalah kelas untuk membuat model Support Vector Regression.
* **regressor = SVR(kernel='rbf'):**
  * Membuat objek SVR dengan kernel RBF dan menyimpannya ke dalam variabel regressor.
  * Kernel: Kernel adalah fungsi yang digunakan untuk memetakan data ke dimensi yang lebih tinggi, di mana SVR dapat menemukan hyperplane yang optimal.
   RBF (Radial Basis Function): Kernel yang umum digunakan dalam SVR. RBF memetakan data ke ruang fitur tak terbatas.
* **regressor.fit(X, y):**
  * Melatih model SVR pada data yang telah disiapkan (X dan y).
  * **fit():** Metode untuk melatih model. Metode ini akan mengoptimalkan parameter model agar sesuai dengan data latih.

Jadi, setelah model dilatih regressor dapat digunakan untuk memprediksi nilai variabel dependen (y) untuk nilai variabel independen (X) yang baru.

*   Melakukan prediksi menggunakan model SVR yang telah dilatih.

**Penjelasan:**

Kode tersebut digunakan untuk memprediksi gaji untuk tingkat posisi tertentu menggunakan model Support Vector Regression (SVR) yang telah dilatih sebelumnya.

* **tingkat_posisi_prediksi = np.array([[6.5]]):**

* **gaji_prediksi = regressor.predict(tingkat_posisi_prediksi):**
  * Memprediksi gaji (gaji_prediksi) menggunakan model SVR (regressor) dengan tingkat_posisi_prediksi yang telah diskalakan sebagai input.
* **gaji_prediksi = sc_y.inverse_transform(gaji_prediksi.reshape(-1, 1)):**

**9. Evaluasi Model SVR**

Langkah terakhir adalah melakukan evaluasi model meliputi MAE, MSE dan R-squared

```python
# Evaluasi model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_actual = y
y_pred = regressor.predict(X)

# Menghitung MAE
mae = mean_absolute_error(y_actual, y_pred)

# Menghitung MSE
mse = mean_squared_error(y_actual, y_pred)

# Menghitung RMSE
rmse = np.sqrt(mse)

# Menghitung R-squared
r2 = r2_score(y_actual, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R-squared:", r2)
```

**Penjelasan:**

Kode tersebut bertujuan untuk mengevaluasi kinerja model Support Vector Regression (SVR) yang telah dilatih sebelumnya. Evaluasi dilakukan dengan menghitung beberapa metrik yang mengukur seberapa baik model dalam memprediksi nilai aktual.

  * **mean_absolute_error:** Menghitung Mean Absolute Error (MAE)

* **y_actual:** Menyimpan nilai aktual dari variabel target (gaji) yang telah diskalakan sebelumnya.
* **y_pred:** Memprediksi nilai target menggunakan model SVR (regressor) dengan data fitur (X) yang telah diskalakan, dan menyimpan hasil prediksi.

**3. Hitung Matrik Evaluasi**
* **mae = mean_absolute_error(y_actual, y_pred):** Mengukur rata-rata selisih absolut antara nilai aktual dan prediksi.
* **mse = mean_squared_error(y_actual, y_pred):** Mengukur rata-rata selisih kuadrat antara nilai aktual dan prediksi.
* **rmse = np.sqrt(mse):** Akar kuadrat dari MSE, memberikan gambaran tentang besarnya error dalam satuan yang sama dengan variabel target.
* **r2 = r2_score(y_actual, y_pred):** Mengukur proporsi varians data yang dijelaskan oleh model (0 hingga 1).

5. Latih model pada data latih dan lakukan prediksi pada data uji.

  **Jawab:**

```python
#Training Model
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm).fit()

# Prediksi pada data uji
X_test_sm = sm.add_constant(X_test)
Y_test_pred = lr.predict(X_test_sm)

plt.scatter(y_test, Y_test_pred)
plt.xlabel("Nialai Aktual")
plt.ylabel("Nilai Prediksi")
plt.title("Scatter Plot: Nilai Aktual vs. Nilai Prediksi")
plt.show()
```

**Penjelasan:**

Kode tersebut melatih model linear regression menggunakan library statsmodels dan memvisualisasikan hasil prediksi terhadap nilai aktual.

* **import statsmodels.api as sm:** Mengimpor library statsmodels dan memberikan alias sm untuk mempermudah penggunaan. Statsmodels menyediakan kelas dan fungsi untuk estimasi model statistik dan melakukan pengujian statistik.
* **X_train_sm = sm.add_constant(X_train):** Menambahkan konstanta (intercept) ke data pelatihan X_train. Statsmodels memerlukan konstanta untuk menghitung intercept dalam model linear regression.
* **lr = sm.OLS(y_train, X_train_sm).fit():**
  * **sm.OLS(y_train, X_train_sm):** Membuat model Ordinary Least Squares (OLS), yang merupakan metode umum untuk estimasi model linear regression.
    - **y_train:** Variabel target (dependen).
    - **X_train_sm:** Variabel fitur (independen) dengan konstanta.
  * **.fit():** Melatih model OLS dan mengembalikan objek yang berisi hasil estimasi.
* **X_test_sm = sm.add_constant(X_test):** Menambahkan konstanta ke data pengujian X_test.
* **Y_test_pred = lr.predict(X_test_sm):** Memprediksi nilai target pada data pengujian menggunakan model lr yang telah dilatih.
* **plt.scatter(y_test, Y_test_pred):** Membuat scatter plot untuk memvisualisasikan:
  * **y_test:** Nilai aktual dari variabel target pada data pengujian.
  * **Y_test_pred:** Nilai prediksi dari model.
  * **plt.xlabel("Nialai Aktual") plt.ylabel("Nilai Prediksi") plt.title("Scatter Plot: Nilai Aktual vs. Nilai Prediksi"):** Memberikan label pada sumbu x, sumbu y, dan judul plot.
* **plt.show():** Menampilkan scatter plot.

6. valuasi model dengan menghitung metrik seperti R-squared, MSE, dan MAE. Tampilkan hasil evaluasi.

  **Jawab:**

```python
# Evaluasi model
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Hitung nilai R-squad
from sklearn.metrics import r2_score

mae = mean_absolute_error(y_test, Y_test_pred)

# Menghitung MSE
mse = mean_squared_error(y_test, Y_test_pred)

# Menghitung RMSE
rmse = np.sqrt(mse)

# Menghitung R-squared
r2 = r2_score(y_test, Y_test_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R-squared:", r2)
```

**Penjelasan:**

Kode tersebut mengevaluasi kinerja model linear regression menggunakan beberapa metrik evaluasi.

* **from sklearn.metrics import mean_absolute_error, mean_squared_error:** Mengimpor fungsi mean_absolute_error dan mean_squared_error dari library Scikit-learn. Fungsi-fungsi ini digunakan untuk menghitung metrik evaluasi model regresi.
* **from sklearn.metrics import r2_score:** Mengimpor fungsi r2_score dari library Scikit-learn untuk menghitung nilai R-squared.
* **mae = mean_absolute_error(y_test, Y_test_pred):** Menghitung Mean Absolute Error (MAE). MAE mengukur rata-rata selisih absolut antara nilai aktual dan nilai prediksi.
* **mse = mean_squared_error(y_test, Y_test_pred):** Menghitung Mean Squared Error (MSE). MSE mengukur rata-rata selisih kuadrat antara nilai aktual dan nilai prediksi.
* **rmse = np.sqrt(mse):** Menghitung Root Mean Squared Error (RMSE). RMSE adalah akar kuadrat dari MSE dan memiliki satuan yang sama dengan variabel target.
* **r2 = r2_score(y_test, Y_test_pred):** Menghitung R-squared. R-squared mengukur proporsi varians variabel target yang dapat dijelaskan oleh model.
* **print("MAE:", mae) print("MSE:", mse) print("RMSE:", rmse) print("R-squared:", r2):** Menampilkan nilai metrik evaluasi MAE, MSE, RMSE, dan R-squared.

Langkah 6 Regresi Linier



```python
#Pisahkan variable bebas (X) dan variable taret (y)
#Membuat variabel bebas X dan Y, contoh pengambilan dari analisis korelasi sebelumnya
X = data['Length of Membership']
y = data['Yearly Amount Spent']
```

- X: Variabel independen (bebas), menggunakan Length of Membership. --> x sebagai variable yang memengaruhi
- y: Variabel dependen (target), yaitu Yearly Amount Spent. --> y sebagai variable yang dipengaruhi
- Ini digunakan untuk memahami hubungan antara durasi keanggotaan dan pengeluaran tahunan.

```python
#Bagi data menjadi data latih (70%) dan data uji (30%) menggunakan train_test_split.
#Pembagian data latih dan data uji dengan proporsi 7:3
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

- from sklearn.model_selection import train_test_split adalah import yang digunakan untuk memanggil fungsi train_test_split dari modul sklearn.model_selection, yang digunakan untuk membagi dataset menjadi data latih dan data uji
- train_test_split memastikan bahwa model dilatih pada sebagian data dan diuji pada sebagian data lainnya. Hal ini penting untuk mengukur performa model secara obyektif dan memastikan model tidak overfitting (terlalu cocok terhadap data latih dan kurang mampu memprediksi data baru).
-Tujuan utama dari fungsi ini adalah untuk membagi dataset menjadi dua bagian:
    - Training set (data latih): Dataset yang digunakan untuk melatih model, biasanya sebagian besar dari total data (misalnya 70-80% dari dataset).
    - Test set (data uji): Dataset yang digunakan untuk menguji performa model, biasanya lebih kecil (misalnya 20-30% dari dataset).
- train_size=0.7: 70% dari data digunakan untuk melatih model.
- test_size=0.3: 30% sisanya digunakan untuk menguji model.
- random_state=100: Untuk menjaga konsistensi pembagian data, angka ini digunakan sebagai seed acak.


- X_train: Subset data latih untuk variabel independen.
- X_test: Subset data uji untuk variabel independen.
- y_train: Subset data latih untuk variabel dependen.
- y_test: Subset data uji untuk variabel dependen.

```python
#Lakukan training model regresi linier menggunakan library StatsModels. Tambahkan konstanta (intercept) ke variabel bebas.
# Training model
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm).fit()
```

- statsmodels.api adalah modul dalam StatsModels, yang digunakan untuk analisis statistik. Modul ini menawarkan berbagai alat untuk regresi, termasuk Ordinary Least Squares (OLS) atau Regresi Linier.
- X_train_sm = sm.add_constant(X_train) -->
  - sm.add_constant(X_train) digunakan untuk menambahkan kolom konstanta (bias atau intercept) ke dalam variabel independen (fitur) X_train.
  - Pada regresi linier, memerlukan intercept (titik di mana garis regresi memotong sumbu-y). Fungsi add_constant() menambahkan kolom yang berisi nilai 1 untuk setiap baris dalam dataset X_train, sehingga model dapat menghitung intercept.
  - Jika tidak menambahkan konstanta, model regresi linier akan berasumsi bahwa garis regresi melewati titik asal (0,0), yang seringkali tidak realistis.
- lr = sm.OLS(y_train, X_train_sm).fit()
  - sm.OLS(y_train, X_train_sm): digunakan untuk mendefinisikan model regresi linier menggunakan Ordinary Least Squares (OLS), yaitu metode statistik yang paling umum digunakan untuk meminimalkan jumlah kuadrat dari perbedaan antara nilai observasi dan nilai yang diprediksi oleh model.
    - y_train: Ini adalah target atau variabel dependen (Yearly Amount Spent), yang ingin kita prediksi.
  - X_train_sm: Ini adalah variabel independen (Length of Membership) setelah ditambahkan konstanta yang akan memungkinkan model menghitung intercept.
  - fit(): Setelah mendefinisikan model dengan OLS, fungsi fit() digunakan untuk melatih (memperkirakan parameter) model regresi linier berdasarkan data latih (X_train_sm dan y_train). Proses fit() ini akan menghasilkan nilai-nilai koefisien regresi yang mendeskripsikan hubungan antara variabel independen dan variabel dependen.

- Hasil dari fit() adalah model terlatih lr, yang berisi:
  - Intercept (Konstanta): Titik di mana garis regresi memotong sumbu-y.
  -Koefisien (Slope): Setiap koefisien mendeskripsikan perubahan yang diharapkan dalam variabel dependen (y) untuk setiap peningkatan satu unit dalam variabel independen (X).

```python
# Visualisasikan garis regresi pada data latih.
# Visualisasi garis regresi
plt.scatter(X_train, y_train)
plt.plot(X_train, 265.2483 + 66.3015*X_train, 'r')
plt.show()
```

- plt.scatter(X_train, y_train): Membuat plot pencar (scatter plot) untuk visualisasi data latih.
- plt.plot(): Menggambar garis regresi berdasarkan persamaan regresi yang diperoleh. Dalam hal ini, persamaan regresinya adalah:
  - Intercept: 265.2483
  - Slope (koefisien): 66.3015
- 'r': Menandakan warna garis merah untuk garis regresi.

Grafik yang dihasilkan akan menampilkan data latih sebagai titik-titik dalam scatter plot. Di atasnya, garis merah merepresentasikan garis regresi yang menunjukkan hubungan linear antara durasi keanggotaan dan pengeluaran tahunan.

Langkah 7 Analisis Residual

```python
# Prediksi nilai y_value dari data x yang telah dilatih
y_train_pred = lr.predict(X_train_sm)

# Menghitung residual
res = (y_train - y_train_pred)
```

- lr.predict(X_train_sm)
  - predict() digunakan untuk menghasilkan prediksi dari model regresi linier yang sudah dilatih (lr). Model ini memprediksi nilai target (y_train_pred) berdasarkan variabel independen X_train_sm (yang sudah ditambahkan konstanta).
  - X_train_sm: Variabel independen (Length of Membership) pada data latih yang sudah ditambahkan konstanta (intercept). Ini digunakan oleh model untuk membuat prediksi sesuai dengan koefisien regresi yang dihitung sebelumnya.
  - y_train_pred: Ini adalah nilai prediksi untuk variabel target y_train berdasarkan data latih X_train.

- Residual: Selisih antara nilai aktual dari variabel target (y_train) dengan nilai prediksi dari model (y_train_pred). Residual dapat digunakan untuk mengukur seberapa baik model mampu memprediksi data.
  - res = (y_train - y_train_pred): Ini menghitung residual untuk setiap observasi pada data latih. Residual menunjukkan error atau kesalahan antara nilai aktual dan nilai prediksi.
  - Interpretasi Residual:
    1. Jika residual mendekati 0, artinya prediksi model sangat akurat untuk observasi tersebut.
    2. Jika residual besar (positif atau negatif), artinya prediksi model jauh dari nilai aktual.

```python
# Lakukan prediksi nilai y dari data latih dan hitung residual (selisih antara nilai sebenarnya dan nilai prediksi).

# Histogram residual
fig = plt.figure()
sns.distplot(res, bins=15)
plt.title('Error Terms', fontsize=15)
plt.xlabel('y_train - y_train_pred', fontsize=15)
plt.show()

# Scatter plot residual
plt.scatter(X_train, res)
plt.show()

Visualisasikan residual dalam bentuk histogram dan scatter plot untuk mengevaluasi distribusi dan pola error.
```

- plt.figure() membuat figure kosong yang dapat digunakan untuk menambahkan subplots, grafik, atau visualisasi lainnya
- sns.distplot(res, bins=15): Fungsi ini digunakan untuk membuat histogram dari residual, yang menunjukkan distribusi kesalahan (selisih antara nilai aktual dan nilai prediksi).
  - res: Residual (selisih antara nilai aktual dan prediksi).
  - bins=15: Menentukan jumlah interval (bin) dalam histogram untuk membagi residual.
  - Histogram: Menunjukkan sebaran residual. Jika distribusi residual berbentuk normal (terdistribusi secara simetris di sekitar 0), hal ini mengindikasikan bahwa model memiliki error yang terdistribusi secara acak, yang merupakan ciri dari model regresi yang baik.
- plt.title(): Menambahkan judul pada plot, di sini 'Error Terms', yang menunjukkan bahwa histogram ini menggambarkan distribusi error atau residual.

- plt.xlabel(): Menambahkan label pada sumbu x, yaitu y_train - y_train_pred (selisih antara nilai aktual dan nilai prediksi).

- plt.show(): Menampilkan plot histogram.

Histogram residual ini memungkinkan untuk melihat apakah residual (error) berdistribusi normal. Sebuah distribusi normal pada residual menunjukkan bahwa model cocok dengan data dan tidak ada bias yang signifikan dalam prediksi.

Simetri di sekitar 0: Jika histogram terlihat simetris dengan puncak di sekitar 0, ini merupakan indikasi yang baik bahwa error tersebar secara acak dan model tidak memiliki bias sistematis.

Langkah 8 Prediksi pada Data Uji dan Evaluasi Model

```python
# Prediksi pada data uji
X_test_sm = sm.add_constant(X_test)
y_test_pred = lr.predict(X_test_sm)
```

- sm.add_constant(X_test) digunakan untuk menambahkan konstanta (intercept) pada variabel independen X_test. Penambahan konstanta diperlukan agar model dapat memperkirakan intercept (titik potong pada sumbu y).
  - X_test data uji yang berisi variabel independen (Length of Membership) yang tidak dilihat oleh model saat pelatihan. Dalam regresi, penambahan kolom dengan nilai konstanta diperlukan untuk memperhitungkan intercept saat membuat prediksi.
- lr.predict(X_test_sm): Metode predict() digunakan untuk membuat prediksi nilai target (y_test_pred) berdasarkan model yang sudah dilatih (lr) menggunakan data uji X_test_sm (yang sudah ditambahkan konstanta).
  - y_test_pred: Hasil prediksi dari model regresi linier untuk variabel target ( Yearly Amount Spent) menggunakan data uji X_test.

```python
# Lakukan prediksi pada data uji.

# Hitung nilai R-squared
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, y_test_pred)

# Hitung nilai R-squared untuk mengukur kinerja model pada data uji.
```

- r2_score: Fungsi ini digunakan untuk menghitung koefisien determinasi atau R-squared. Ini adalah metrik yang mengukur seberapa baik model regresi dapat menjelaskan variabilitas dalam data
- r2_score(y_test, y_test_pred):
  - y_test: Nilai aktual dari target pada data uji (Yearly Amount Spent).
  - y_test_pred: Nilai prediksi dari model yang dihasilkan menggunakan data uji.
- R-squared (R): Merupakan ukuran yang menunjukkan proporsi variansi variabel dependen yang dapat dijelaskan oleh variabel independen dalam model.
  - Nilai R berkisar dari 0 hingga 1:
    - R = 1: Model dapat menjelaskan seluruh variasi dalam data dengan sempurna (fit sempurna).
    - R = 0: Model tidak dapat menjelaskan variansi dalam data sama sekali.
    - Nilai R negatif menunjukkan bahwa model lebih buruk daripada model garis rata-rata.

  - StandardScaler --> kelas dari scikit-learn yang digunakan untuk melakukan standardisasi pada data. Standardisasi mengubah data sehingga memiliki mean 0 dan standard deviation 1.
- sc_X = StandardScaler()
  - sc_X merupakan Objek StandardScaler yang akan digunakan untuk menstandardisasi variabel independen X.
- sc_y = StandardScaler()
  - sc_y: Objek StandardScaler yang akan digunakan untuk menstandardisasi variabel target y.
- X = sc_X.fit_transform(X.reshape(-1, 1))
  - fit_transform(): Kombinasi dari fit (untuk menghitung mean dan standard deviation) dan transform (untuk mengubah data berdasarkan nilai yang dihitung).
  - X.reshape(-1, 1). Karena X adalah array 1D, fungsi reshape(-1, 1) mengubahnya menjadi array 2D dengan satu kolom. Diperlukan karena StandardScaler mengharapkan input dalam bentuk 2D array.
- y = sc_y.fit_transform(y.reshape(-1, 1))
  - fit_transform(): Kombinasi dari fit (untuk menghitung mean dan standard deviation) dan transform (untuk mengubah data berdasarkan nilai yang dihitung).
  - Y.reshape(-1, 1). Karena Y adalah array 1D, fungsi reshape(-1, 1) mengubahnya menjadi array 2D dengan satu kolom. Diperlukan karena StandardScaler mengharapkan input dalam bentuk 2D array.
- Setelah data distandardisasi, X dan y akan berubah dalam bentuk data terstandardisasi (dengan mean 0 dan standar deviasi 1).


Fitting SVR ke Dataset

Lakukan pembuatan model SVR dengan kernel RBF (Radial Basis Function) dan melatihnya dengan data yang telah di-scaled.

```python
# Fitting SVR ke dataset
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(X, y)
```

- digunakan untuk melakukan fit (melatih) model Support Vector Regression (SVR) pada dataset yang telah di-feature scale menggunakan kernel radial basis function (RBF).
- from sklearn.svm import SVR
    - SVR: Support Vector Regression adalah versi regresi dari Support Vector Machine (SVM). SVR bertujuan untuk menemukan hyperplane yang paling baik dalam merepresentasikan hubungan antara variabel independen (X) dan target (y) dengan memperhatikan margin error.
- regressor = SVR(kernel='rbf')
  - kernel='rbf'--> menunjukkan bahwa kernel Radial Basis Function (RBF) digunakan dalam SVR. Kernel ini sangat umum digunakan untuk menangani masalah non-linear, karena kernel RBF mampu memetakan data dari dimensi rendah ke dimensi yang lebih tinggi untuk menemukan pola yang lebih kompleks.
    - RBF mengubah hubungan non-linear menjadi lebih mudah dimodelkan dalam ruang dimensi yang lebih tinggi.
  - regressor --> adalah objek model SVR yang siap untuk dilatih menggunakan data.
- regressor.fit(X, y)
  - fit(X, y) digunakan untuk melatih model SVR dengan data yang telah distandardisasi (baik variabel independen X maupun target y sudah di-scaled menggunakan StandardScaler).
    - X: Data fitur (variabel independen) yang telah di-scaled.
    - y: Target (variabel dependen) yang juga telah di-scaled.

Prediksi Hasil

- Membuat array 2D yang berisi tingkat posisi yang akan diprediksi. Dalam contoh ini, tingkat posisi 6.5.

- Menskalakan fitur prediksi menggunakan sc_X.transform.

- Melakukan prediksi menggunakan model SVR yang telah dilatih.

- Mengembalikan hasil prediksi ke dalam skala aslinya menggunakan sc_y.inverse_transform.

```python
# Prediksi hasil
# Buat array 2D yang berisi tingkat posisi yang akan diprediksi
tingkat_posisi_prediksi = np.array([[6.5]])
# Penskalaan fitur untuk data yang akan diprediksi
tingkat_posisi_prediksi = sc_X.transform(tingkat_posisi_prediksi)
# Melakukan prediksi menggunakan model SVR
gaji_prediksi = regressor.predict(tingkat_posisi_prediksi)
# Kembalikan hasil prediksi ke skala aslinya
gaji_prediksi = sc_y.inverse_transform(gaji_prediksi.reshape(-1, 1))
```

- tingkat_posisi_prediksi = np.array([[6.5]]) --> np.array([[6.5]]): Membuat array 2D berisi nilai 6.5, yang merepresentasikan tingkat posisi yang ingin diprediksi.
- tingkat_posisi_prediksi = sc_X.transform(tingkat_posisi_prediksi) --> Menggunakan sc_X, objek StandardScaler yang sebelumnya telah dilatih dengan data X, untuk menstandardisasi tingkat posisi yang baru (6.5).
- gaji_prediksi = regressor.predict(tingkat_posisi_prediksi) --> Menggunakan model SVR yang telah dilatih untuk memprediksi gaji berdasarkan nilai tingkat posisi yang sudah di-scale (hasil dari langkah sebelumnya). Hasil prediksi ini masih dalam skala yang distandardisasi (karena model dilatih pada data yang sudah di-scale).
- gaji_prediksi = sc_y.inverse_transform(gaji_prediksi.reshape(-1, 1)) --> digunakan untuk mengubah hasil prediksi dari skala yang distandardisasi kembali ke skala aslinya (gaji dalam satuan sebenarnya).
gaji_prediksi.reshape(-1, 1): Sebelum melakukan inverse transform, hasil prediksi (yang masih dalam bentuk 1D array) diubah menjadi array 2D dengan satu kolom, karena inverse_transform() mengharapkan input berupa array 2D.

Evaluasi Model SVR

```python
# Evaluasi model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_actual = y
y_pred = regressor.predict(X)

# Menghitung MAE
mae = mean_absolute_error(y_actual, y_pred)

# Menghitung MSE
mse = mean_squared_error(y_actual, y_pred)

# Menghitung RMSE
rmse = np.sqrt(mse)

# Menghitung R-squared
r2 = r2_score(y_actual, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R-squared:", r2)
```

- from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
  - mean_absolute_error: Menghitung Mean Absolute Error (MAE), yaitu rata-rata dari nilai absolut kesalahan prediksi.
  - mean_squared_error: Menghitung Mean Squared Error (MSE), yaitu rata-rata dari kuadrat kesalahan prediksi.
  - r2_score: Menghitung R-squared atau koefisien determinasi, yang menunjukkan seberapa baik model menjelaskan variasi dalam data target.

- y_actual = y -->  Menyimpan nilai target asli dari dataset dalam variabel y_actual untuk digunakan dalam perhitungan evaluasi. y adalah array yang berisi nilai target asli.
- y_pred = regressor.predict(X) --> Menggunakan model SVR (regressor) untuk memprediksi nilai target dari data fitur X. y_pred adalah array dari nilai prediksi yang dihasilkan oleh model untuk data fitur X.
- mae = mean_absolute_error(y_actual, y_pred) --> Menghitung Mean Absolute Error (MAE) antara nilai target asli (y_actual) dan nilai prediksi (y_pred). mae memberikan rata-rata dari selisih absolut antara nilai yang diprediksi dan nilai aktual, mengukur seberapa besar kesalahan rata-rata.
- mse = mean_squared_error(y_actual, y_pred) --> Menghitung Mean Squared Error (MSE) antara nilai target asli dan nilai prediksi. mse memberikan rata-rata dari kuadrat selisih antara nilai yang diprediksi dan nilai aktual, mengukur seberapa besar kesalahan dalam unit kuadrat.
- rmse = np.sqrt(mse) --> Menghitung Root Mean Squared Error (RMSE) dengan mengambil akar kuadrat dari MSE. rmse memberikan ukuran kesalahan dalam satuan yang sama dengan data target, memberikan gambaran yang lebih intuitif tentang seberapa besar kesalahan prediksi.
- r2 = r2_score(y_actual, y_pred) --> Menghitung nilai R-squared (R), yang menunjukkan proporsi variasi dalam nilai target yang dapat dijelaskan oleh model. r2 memberikan indikasi seberapa baik model cocok dengan data. Nilai mendekati 1 menunjukkan bahwa model menjelaskan sebagian besar variasi dalam data.
- print("MAE:", mae): Menampilkan nilai Mean Absolute Error (MAE) yang dihitung.
- print("MSE:", mse): Menampilkan nilai Mean Squared Error (MSE) yang dihitung.
- print("RMSE:", rmse): Menampilkan nilai Root Mean Squared Error (RMSE) yang dihitung.
- print("R-squared:", r2): Menampilkan nilai R-squared (R) yang dihitung.

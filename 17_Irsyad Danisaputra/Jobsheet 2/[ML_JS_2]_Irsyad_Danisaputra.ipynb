```python
import pandas as pd

# df = pd.read_csv('data/Titanic-Dataset-selected.csv')

df = pd.read_csv(io.BytesIO(uploaded['Titanic-Dataset-selected.csv'])) # load dataset
df.head()
```

## Langkah 2 - Split Data

```python
# Split data
from sklearn.model_selection import train_test_split

# Split data training dan dan lainnya
# data lainnya, akan kita split lagi menjadi validasi dan testing.
# Rasio yang akan kita gunakan adalah 8:1:1
df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0)

# Split lagi antara validasi dan testing
df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0)

# Cek masing-masing ukuran data

print(f'Jumlah data asli: {df.shape[0]}')
print(f'Jumlah data train: {df_train.shape[0]}')
print(f'Jumlah data val: {df_val.shape[0]}')
print(f'Jumlah data test: {df_test.shape[0]}')

# Cek rasio tiap label
print('=========')
print(f'Jumlah label data asli:\n{df.Survived.value_counts()}')
print(f'Jumlah label data train:\n{df_train.Survived.value_counts()}')
print(f'Jumlah label data val:\n{df_val.Survived.value_counts()}')
print(f'Jumlah label data test:\n{df_test.Survived.value_counts()}')
```

**Penjelasan**
<br>- df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0)
<br> Membagi dataset menjadi training data dan unseen data dengan rasio 8:1:1

<br> - df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0)
<br> Membagi df_unseen menjadi validasi dan testing data masing-masing 50%

```python
# import pandas as pd
# import io

# df_tugas = pd.read_csv(io.BytesIO(uploaded['wbc.csv'])) # load dataset

import pandas as pd
import io

df_tugas = pd.read_csv(io.BytesIO(uploaded['wbc (1).csv'])) # load dataset using the actual key 'wbc (1).csv'
```

```python
df_tugas.info()
df_tugas.head()
```

### 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
df_tugas = df_tugas.drop(['id','Unnamed: 32'], axis=1)
df_tugas.head()
```

**Penjelasan:** perintah ini digunakan menghapus kolom id dan Unnamed: 32 dari DataFrame df_tugas dan kemudian menampilkan lima baris pertama dari DataFrame yang telah diperbarui.

### 2. Lakukan proses encoding pada kolom "diagnosis".

```python
from sklearn.preprocessing import LabelEncoder

# Inisialisasi LabelEncoder
le = LabelEncoder()

# Lakukan encoding pada kolom 'diagnosis'
df_tugas['diagnosis'] = le.fit_transform(df_tugas['diagnosis'])

df_tugas.head()
```

**Penjelasan:** perintah ini digunakan untuk mengonversi label kategori dalam kolom 'diagnosis' menjadi angka dan menampilkan hasilnya.

### 3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance
std = StandardScaler()

# Pilih semua kolom numerik
numerical_columns = df_tugas.select_dtypes(include=['float64', 'int64']).columns

# Lakukan standarisasi pada kolom-kolom numerik
df_tugas[numerical_columns] = std.fit_transform(df_tugas[numerical_columns])

# Tampilkan DataFrame setelah standarisasi
df_tugas.head()
```

**Penjelasan:** perintah ini menstandarisasi semua kolom numerik dalam DataFrame dan menampilkan hasilnya.

### 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
from sklearn.model_selection import train_test_split

# kolom diagnosis perlu distratify
df_tugas_train, df_tugas_test = train_test_split(df_tugas, test_size=0.2, random_state=0, stratify=df_tugas['diagnosis'])

print(f'Jumlah label data asli:\n{df_tugas.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_tugas_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_tugas_test.diagnosis.value_counts()}')
```

**Penjelasan:** kode di atas berfungsi untuk membagi data menjadi set pelatihan dan pengujian sambil memastikan distribusi label yang konsisten di kedua set, lalu mencetak jumlah label di masing-masing set.

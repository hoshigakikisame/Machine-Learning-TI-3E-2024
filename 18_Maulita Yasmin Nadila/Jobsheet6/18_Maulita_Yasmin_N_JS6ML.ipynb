Import Library

* import pandas as pd: mengimpor library Pandas, yang digunakan untuk manipulasi dan analisis data, terutama data berbentuk tabel (DataFrames).

* from sklearn.model_selection import train_test_split: mengimpor fungsi train_test_split dari sklearn. Fungsi ini digunakan untuk membagi dataset menjadi dua bagian: data pelatihan (training set) dan data pengujian (testing set).

Persiapan Data

```python
# Load data
df = pd.read_csv('/content/drive/MyDrive/Mesin Learning/wbc (1).csv')

df.head()
```

* df = pd.read_csv('/content/drive/MyDrive/Mesin Learning/wbc (1).csv'): digunakan untuk membaca file CSV (Comma Separated Values) ke dalam DataFrame menggunakan Pandas.
* df.head(): digunakan untuk melihat 5 baris pertama dari DataFrame df.

```python
# Cek kolom null
df.isnull().sum()
```

kode diatas digunakan untuk mengecek apakah terdapat missing values (nilai yang hilang atau kosong) di dalam setiap kolom pada DataFrame df, kemudian terdapat method sum() yang diterapkan setelah isnull() yang berguna untuk menjumlahkan nilai true setiap kolom.

```python
# Seleksi fitur

# Slice dataframe mulai dari kolom 'radius_mean' sampai 'fractal_dimension_worst'
X = df.iloc[:,3:-1]
y = df['diagnosis']
y = y.map({'M':1, 'B':0}) # Encode label

# Cek jumlah fitur dan instance
X.shape
```

* Seleksi fitur (X) : Data yang disimpan dalam variabel X digunakan untuk melakukan indexing berbasis posisi
* Seleksi Label (y) : Data yang disimpan dalam y digunakan untuk menyimpan kolom diagnosis (label target) dan untuk mengubah nilai-nilai kategoris dalam kolom y menjadi numerik.

Split data training dan testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

* from sklearn.model_selection import train_test_split:
Mengimpor fungsi train_test_split dari modul sklearn.model_selection.
* X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1):
1. X_train dan X_test:
Ini adalah hasil pemisahan dari fitur X (data) ke dalam dua bagian. X_train adalah data yang digunakan untuk melatih model (80% dari data asli), dan X_test adalah data yang digunakan untuk menguji model (20% dari data asli).
2. y_train dan y_test:
Ini adalah hasil pemisahan dari y (label atau target) menjadi dua bagian yang sesuai dengan X_train dan X_test. y_train digunakan untuk melatih model, dan y_test digunakan untuk evaluasi performa model.
3. test_size=0.2:
Parameter ini mengatur proporsi data uji. Dalam hal ini, 20% dari data digunakan sebagai data pengujian, sementara sisanya 80% digunakan sebagai data pelatihan.
4. random_state=1:
Parameter ini digunakan untuk mengatur seed dari pembagian dataset secara acak, sehingga pembagian dataset akan konsisten jika kode dijalankan kembali. Dengan random_state=1, hasil pembagian akan sama setiap kali kode dijalankan.


Traning Decision Tree

```python
# Secara default, DecisionTreeClassifier dari scikit-learn akan menggunakan nilai "Gini" untuk kriteria
# Terdapat beberapa "hyperparamater" yang dapat digunakan. Silahka baca dokumentasi
# Pada kasus ini kita akan menggunakan parameter default
dt = DecisionTreeClassifier()

# Sesuaikan dt ke set training
dt.fit(X_train, y_train)

# Memprediksi label set test
y_pred_dt = dt.predict(X_test)

#  menghitung set accuracy
acc_dt = accuracy_score(y_test, y_pred_dt)
print("Test set accuracy: {:.2f}".format(acc_dt))
print(f"Test set accuracy: {acc_dt}")
```

* DecisionTreeClassifier() menginisialisasi model Decision Tree dari Scikit-learn.
* dt.fit(X_train, y_train):
fit() adalah metode yang digunakan untuk melatih model menggunakan data pelatihan.
X_train adalah data fitur pelatihan, dan y_train adalah label atau target yang sesuai.
* y_pred_dt = dt.predict(X_test):
predict() digunakan untuk memprediksi label (klasifikasi) dari data uji X_test.
* acc_dt = accuracy_score(y_test, y_pred_dt):
accuracy_score() adalah fungsi dari Scikit-learn yang digunakan untuk menghitung akurasi model. y_test adalah label sebenarnya, dan y_pred_dt adalah label yang diprediksi oleh model.
* print("Test set accuracy: {:.2f}".format(acc_dt)):
Baris ini mencetak hasil akurasi model dalam format dua desimal.
* print(f"Test set accuracy: {acc_dt}"):
Alternatif lain untuk mencetak akurasi dengan f-string, yang lebih mudah dibaca dan fleksibel.

Training RandomForest

```python
# Pada kasus kali ini kita akan menggunakan estimator pada RandomForest
# Untuk detail parameter (hyperparameter) silahkan cek dokumentasi

rf = RandomForestClassifier(n_estimators=10, random_state=1)

# Sesuaikan dt ke set training
rf.fit(X_train, y_train)

# Memprediksi label set test
y_pred_rf = rf.predict(X_test)

#  menghitung set accuracy
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Test set accuracy: {:.2f}".format(acc_rf))
print(f"Test set accuracy: {acc_rf}")
```

* RandomForestClassifier() adalah model ensemble dari Scikit-learn yang membangun banyak Decision Trees dan menggabungkan prediksinya (melalui voting) untuk meningkatkan akurasi.
* n_estimators=10: Model akan membangun 10 pohon keputusan (Decision Trees). Semakin banyak pohon yang digunakan, biasanya hasilnya semakin stabil, namun dengan biaya waktu komputasi yang lebih tinggi.
* random_state=1: Ini digunakan untuk membuat hasil yang konsisten setiap kali kode dijalankan, dengan mengatur seed acak untuk pembagian data dan penentuan pohon.
* fit() adalah metode yang digunakan untuk melatih model pada data pelatihan.
* X_train adalah fitur dari data pelatihan, dan y_train adalah label atau target yang sesuai.
* predict() digunakan untuk memprediksi label (klasifikasi) dari data uji X_test.
* accuracy_score() digunakan untuk menghitung akurasi dari model. Akurasi dihitung sebagai persentase dari prediksi yang benar terhadap keseluruhan data uji. y_test adalah label sebenarnya, dan y_pred_rf adalah label yang diprediksi oleh model Random Forest.
* print("Test set accuracy: {:.2f}".format(acc_rf)): Baris ini mencetak hasil akurasi model dalam format dua desimal.
* print(f"Test set accuracy: {acc_rf}"): alternatif menggunakan f-string untuk mencetak akurasi.

* Seleksi fitur (X) : Data yang disimpan dalam variabel X digunakan untuk melakukan indexing berbasis posisi

```python
layer_one_estimators = [
                        ('rf_1', RandomForestClassifier(n_estimators=10, random_state=42)),
                        ('knn_1', KNeighborsClassifier(n_neighbors=5))
                       ]
layer_two_estimators = [
                        ('dt_2', DecisionTreeClassifier()),
                        ('rf_2', RandomForestClassifier(n_estimators=50, random_state=42)),
                       ]
layer_two = StackingClassifier(estimators=layer_two_estimators, final_estimator=LogisticRegression())


clf = StackingClassifier(estimators=layer_one_estimators, final_estimator=layer_two)

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
clf.fit(X_train, y_train).score(X_test, y_test)
```

* layer_one_estimators: Ini adalah list yang berisi dua model yang akan digunakan sebagai estimators untuk lapisan pertama dalam model stacking.
* ('rf_1', RandomForestClassifier(n_estimators=10, random_state=42)):
Membuat model Random Forest dengan 10 estimators (pohon keputusan) dan menggunakan random_state=42 untuk memastikan hasil yang konsisten setiap kali kode dijalankan.
* ('knn_1', KNeighborsClassifier(n_neighbors=5)):
Membuat model K-Nearest Neighbors (KNN) dengan parameter n_neighbors=5, yang berarti model ini akan mempertimbangkan 5 tetangga terdekat untuk prediksi.
* layer_two_estimators: Ini adalah list yang berisi dua model untuk lapisan kedua dalam model stacking.
* ('dt_2', DecisionTreeClassifier()):
Membuat model Decision Tree menggunakan parameter default (kriteria "Gini").
* ('rf_2', RandomForestClassifier(n_estimators=50, random_state=42)):
Membuat model Random Forest dengan 50 estimators dan random_state=42, memberikan hasil yang konsisten.
* layer_two: Ini adalah model stacking yang dibangun dengan menggunakan estimators dari layer_two_estimators.
* StackingClassifier: Menggabungkan estimators yang ada untuk memprediksi hasil.
* final_estimator=LogisticRegression(): Model akhir dari stacking adalah Logistic Regression, yang akan mengambil output dari layer_two_estimators sebagai input untuk prediksi akhir.
* clf: Ini adalah model stacking yang dibangun dengan menggunakan estimators dari layer_one_estimators.
* StackingClassifier: Model ini menggabungkan output dari layer_one_estimators dan menggunakan layer_two sebagai final estimator.
* train_test_split: Fungsi ini membagi dataset ke dalam data pelatihan dan data pengujian.
* clf.fit(X_train, y_train): Melatih model stacking clf menggunakan data pelatihan X_train dan label y_train.
* .score(X_test, y_test): Menghitung dan mengembalikan akurasi model pada data pengujian X_test dan label y_test.

```python
# Membuat model SVM dengan kernel polynomial

# SVC digunakan untuk membangun model klasifikasi menggunakan Support Vector Machine
svm_model = SVC(kernel='poly', degree=3, random_state=42)
```

```python
# Membuat model Decision Tree

# DecisionTreeClassifier digunakan untuk membangun model klasifikasi menggunakan algoritma pohon keputusan
decision_tree_model = DecisionTreeClassifier(random_state=42)
```

```python
# Menggabungkan ketiga model dalam ensemble voting

# VotingClassifier digunakan untuk menggabungkan beberapa model klasifikasi menjadi satu
ensemble_model = VotingClassifier(
    estimators=[('lr', logistic_model), ('svm', svm_model), ('dt', decision_tree_model)],
    voting='hard'
)
```

```python
# Melatih model ensemble Voting Classifier menggunakan data pelatihan
ensemble_model.fit(X_train, y_train)
```

```python
# Menguji model ensemble voting

# Menggunakan model ensemble untuk memprediksi kelas pada data pengujian
y_pred_ensemble = ensemble_model.predict(X_test)

# Menghitung akurasi prediksi dengan membandingkan label aktual dengan prediksi
accuracy_ensemble = accuracy_score(y_test, y_pred_ensemble)
```

```python
# Menampilkan akurasi model ensemble voting
print("Akurasi Ensemble Voting: {:.2f}%".format(accuracy_ensemble * 100))
```

## Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
diameter_pizza = np.array([6, 8, 10, 14, 18])
harga_pizza = np.array([7, 9, 13, 17.5, 18])

# Estimasi parameter
beta_0 = 1.82
beta_1 = 0.974

# Model regresi
harga_prediksi = beta_0 + beta_1 * diameter_pizza

# Membuat plot
plt.scatter(diameter_pizza, harga_pizza, label='Data Asli')
plt.plot(diameter_pizza, harga_prediksi, color='red', label='Model Regresi')
plt.xlabel('Diameter Pizza (inci)')
plt.ylabel('Harga Pizza (dolar)')
plt.title('Simple Linear Regression: Hubungan Diameter Pizza dengan Harga Pizza')
plt.legend()
plt.grid(True)
plt.show()
```

* plt.plot(diameter_pizza, harga_prediksi, color='red', label='Model Regresi'): Menambahkan garis regresi linier ke plot dengan warna merah dan memberi label 'Model Regresi'.
* plt.xlabel('Diameter Pizza (inci)'): Menambahkan label untuk sumbu X.
* plt.ylabel('Harga Pizza (dolar)'): Menambahkan label untuk sumbu Y.
* plt.title('Simple Linear Regression: Hubungan Diameter Pizza dengan Harga Pizza'): Menambahkan judul pada plot.

## Evaluasi Model dengan Cost Function

```python
import numpy as np

# Data
harga_pizza_asli = np.array([7, 9, 13, 17.5, 18])
harga_pizza_prediksi = np.array([1.82 + 0.974 * 6, 1.82 + 0.974 * 8, 1.82 + 0.974 * 10, 1.82 + 0.974 * 14, 1.82 + 0.974 * 18])

# Menghitung RSS
RSS = np.sum((harga_pizza_asli - harga_pizza_prediksi) ** 2)
print("RSS:", RSS)
```

* harga_pizza_asli = np.array([7, 9, 13, 17.5, 18]): Mendefinisikan array NumPy harga_pizza_asli yang berisi harga pizza aktual.
* harga_pizza_prediksi = np.array([1.82 + 0.974 * 6, 1.82 + 0.974 * 8, 1.82 + 0.974 * 10, 1.82 + 0.974 * 14, 1.82 + 0.974 * 18]): Menghitung prediksi harga pizza untuk diameter yang sama dengan menggunakan model regresi linier:
 - 1.82 adalah intercept (beta_0).
 - 0.974 adalah slope (beta_1).
 - Menghitung harga prediksi untuk diameter pizza 6, 8, 10, 14, dan 18.
* RSS = np.sum((harga_pizza_asli - harga_pizza_prediksi) ** 2):
 - Menghitung Residual Sum of Squares (RSS):
   * harga_pizza_asli - harga_pizza_prediksi: Selisih antara harga aktual dan prediksi (residual).
   * (harga_pizza_asli - harga_pizza_prediksi) ** 2: Menghitung kuadrat dari residual.
   * np.sum(): Menjumlahkan semua kuadrat residual untuk mendapatkan RSS.
* print("RSS:", RSS): Menampilkan nilai RSS.

## Evaluasi Model dengan Cost FunctionEvaluasi Model Dengan R-squared (R2)

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Data
harga_pizza_asli = np.array([7, 9, 13, 17.5, 18])
diameter_pizza = np.array([6, 8, 10, 14, 18])

# Estimasi parameter
beta_0 = 1.82
beta_1 = 0.974

# Prediksi harga pizza
harga_pizza_prediksi = beta_0 + beta_1 * diameter_pizza

# Menghitung RSS (Residual Sum of Squares)
RSS = np.sum((harga_pizza_asli - harga_pizza_prediksi) ** 2)

# Menghitung TSS (Total Sum of Squares)
TSS = np.sum((harga_pizza_asli - np.mean(harga_pizza_asli)) ** 2)

# Menghitung R-squared (R2)
R2 = 1 - (RSS / TSS)

# Menampilkan predicted price in dollar
print("Predicted Price in Dollar:", harga_pizza_prediksi)

# Menghitung MAE (Mean Absolute Error)
mae = mean_absolute_error(harga_pizza_asli, harga_pizza_prediksi)

# Menghitung MSE (Mean Squared Error)
mse = mean_squared_error(harga_pizza_asli, harga_pizza_prediksi)

print("Residual Sum of Squares (RSS)", RSS)
print("Total Sum of Squares (TSS)", TSS)
print("R-squared (R2):", R2)
print("MAE (Mean Absolute Error):", mae)
print("MSE (Mean Squared Error):", mse)
```

Hasil:
- **Predicted Price in Dollar**: Harga pizza yang diprediksi berdasarkan diameter.
- **RSS**: Total kesalahan kuadrat.
- **TSS**: Total variasi dalam data.
- **R**: Proporsi variasi yang dijelaskan oleh model.
- **MAE**: Rata-rata kesalahan absolut.
- **MSE**: Rata-rata kesalahan kuadrat.

Metrik ini membantu dalam mengevaluasi seberapa baik model regresi linier sesuai dengan data aktual.

## Multiple Linear Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Data training
X_train = np.array([
    [6, 2],
    [8, 1],
    [10, 0],
    [14, 2],
    [18, 3]
])
y_train = np.array([7, 9, 13, 17.5, 18])

# Data testing
X_test = np.array([
    [8, 2],
    [9, 0],
    [11, 2],
    [16, 2],
    [12, 0]
])
y_test = np.array([11, 8.5, 15, 18, 11])

# Inisialisasi model Linear Regression
model = LinearRegression()

# Melatih model dengan data training
model.fit(X_train, y_train)

# Melakukan prediksi dengan data testing
y_pred = model.predict(X_test)

# Menghitung R-squared (R2) dari model terlatih
r2 = r2_score(y_test, y_pred)

# Menampilkan hasil prediksi dan nilai target
print("Hasil Prediksi Harga Pizza:")
for i in range(len(y_pred)):
    print(f"Data Test {i + 1}: Prediksi ${y_pred[i]:.2f}, Target ${y_test[i]:.2f}")

# Menampilkan nilai R-squared
print(f"\nR-squared (R2): {r2:.2f}")
```

1. Mengimpor NumPy untuk manipulasi data, LinearRegression dari Scikit-learn untuk membuat model regresi linier, dan r2_score untuk menghitung metrik evaluasi.
2. X_train: Data fitur untuk pelatihan, dengan dua fitur (misalnya, diameter pizza dan jumlah topping). y_train: Target (harga pizza) untuk pelatihan.
3. X_test: Data fitur untuk pengujian. y_test: Target untuk pengujian.
4. Inisialisasi Model Linear Regression: model = LinearRegression()
5. Melatih Model: model.fit(X_train, y_train)
6. Menggunakan model terlatih untuk memprediksi harga pizza berdasarkan data fitur pengujian. y_pred = model.predict(X_test)
7. Menghitung R untuk mengevaluasi seberapa baik model sesuai dengan data pengujian. r2 = r2_score(y_test, y_pred)
8. Menampilkan prediksi dan target aktual untuk setiap data pengujian. print("Hasil Prediksi Harga Pizza:")
for i in range(len(y_pred)): print(f"Data Test {i + 1}: Prediksi ${y_pred[i]:.2f}, Target ${y_test[i]:.2f}")
9. Menampilkan nilai R untuk mengevaluasi performa model. print(f"\nR-squared (R2): {r2:.2f}")




## Regresi Polynomial

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Data training
X_train = np.array([
    [6, 2],
    [8, 1],
    [10, 0],
    [14, 2],
    [18, 3]
])
y_train = np.array([7, 9, 13, 17.5, 18])

# Data testing
X_test = np.array([
    [8, 2],
    [9, 0],
    [11, 2],
    [16, 2],
    [12, 0]
])
y_test = np.array([11, 8.5, 15, 18, 11])

# Inisialisasi model Linear Regression
model = LinearRegression()

# Membentuk polinomial fitur
poly = PolynomialFeatures(degree=2)  # Ganti degree sesuai dengan derajat polinomial yang diinginkan
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Melatih model dengan data training
model.fit(X_train_poly, y_train)

# Melakukan prediksi dengan data testing
y_pred = model.predict(X_test_poly)

# Menghitung R-squared (R2) dari model terlatih
r2 = r2_score(y_test, y_pred)

# Plot hasil regresi
plt.scatter(X_test[:, 0], y_test, color='blue', label='Data Testing')
plt.scatter(X_test[:, 0], y_pred, color='red', label='Prediksi')
plt.title('Grafik Polynomial Regression')
plt.xlabel('Diameter Pizza (inci)')
plt.ylabel('Harga Pizza (dolar)')
plt.legend()
plt.show()

# Menampilkan nilai R-squared
print(f"R-squared (R2): {r2:.2f}")
```

1. Impor Library: NumPy untuk manipulasi data. Matplotlib untuk visualisasi data. PolynomialFeatures untuk membuat fitur polinomial. LinearRegression untuk membuat model regresi. r2_score untuk mengevaluasi model.
2. Data Training: X_train: Fitur pelatihan dengan dua fitur (misalnya, diameter pizza dan jumlah topping). y_train: Target pelatihan (harga pizza).
3. Data Testing: X_test: Fitur pengujian. y_test: Target pengujian.
4. Inisialisasi Model Linear Regression: model = LinearRegression()
5. Membentuk Fitur Polinomial: PolynomialFeatures(degree=2): Membuat fitur polinomial derajat 2. fit_transform: Mengubah fitur pelatihan menjadi fitur polinomial. transform: Mengubah fitur pengujian menjadi fitur polinomial dengan derajat yang sama
6. Melatih Model: model.fit(X_train_poly, y_train)
7. Melakukan Prediksi: y_pred = model.predict(X_test_poly)
8. Menghitung R-squared (R): r2 = r2_score(y_test, y_pred)
9. Plot Hasil Regresi: Menampilkan grafik dengan data pengujian (berwarna biru) dan hasil prediksi (berwarna merah).
10. Menampilkan Nilai R-squared: print(f"R-squared (R2): {r2:.2f}")
